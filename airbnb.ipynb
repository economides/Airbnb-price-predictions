```python
import numpy as np
import pandas as pd
import os
import pathlib


#I have extracted the archive.zip file to the file named archive
#Inside the archive/ folder there is the AB_US_2020.csv file - dataset used for this task
path = "archive/"

airbnb = pd.read_csv(path + "AB_US_2020.csv")
```

```python
def dataframe_info(df):
    """
    Getting the shape of the dataframe & other attributes.
    """
    print("The dataframe has {} rows and {} columns".format(df.shape[0], df.shape[1]))
    print("\n")
    print(df.info())
    print("\n")
    print(df.describe(include="all"))

    return df

def dataframe_sample(df, rows = 10):
    """
    Returning a sample of 10 rows of the dataframe.
    """
    return df.sample(rows)

    





#using the pipe method - in oder to apply the functions in an easily read sequence.

# (airbnb.pipe(dataframe_info).pipe(dataframe_sample))
dataframe_info(airbnb)
print("\nSample print of the airbnb dataframe\n")
dataframe_sample(airbnb)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


print("Category Name count")

plt.figure(figsize = (10,12))

sns.set_context("paper", font_scale=1.5)
g = sns.countplot(y='neighbourhood_group', data=airbnb, order = airbnb['neighbourhood_group'].value_counts().index,palette="tab10")
g.set_yticklabels(g.get_yticklabels(),rotation=30)
g.set_title("Number listings per Neighbourhood Group", fontsize=16)
g.set_xlabel("Number of listings", fontsize=14)
g.set_ylabel("Neighbourhood Group Names", fontsize=14)
plt.show()
```
